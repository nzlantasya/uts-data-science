```python
import matplotlib.pyplot as plt

bulan = ["Jan", "Feb", "Mar", "Apr"]
penjualan = [100, 120, 150, 180]

plt.plot(bulan, penjualan)
plt.title("Penjualan Bulanan")
plt.show()
```

```python
plt.bar(bulan, penjualan)
plt.title("Bar Chart Penjualan")
plt.show()
```

## Kesimpulan

Visualisasi data mempermudah pemahaman pola dan tren dalam data.
