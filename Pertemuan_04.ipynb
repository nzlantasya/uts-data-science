```python
import pandas as pd

data = [75, 80, 85, 90, 95]

s = pd.Series(data)

print("Mean :", s.mean())
print("Median :", s.median())
print("Standar Deviasi :", s.std())
```

```python
s.describe()
```

## Kesimpulan

Statistika dasar membantu memahami karakteristik data melalui mean, median, dan standar deviasi.
