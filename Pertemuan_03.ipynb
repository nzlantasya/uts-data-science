```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Nama": ["A", "B", "C", "D"],
    "Nilai": [80, np.nan, 90, 85]
})

df
```

```python
df["Nilai"] = df["Nilai"].fillna(df["Nilai"].mean())
df
```

## Kesimpulan

Data cleaning diperlukan untuk menangani missing value sehingga data lebih siap dianalisis.
