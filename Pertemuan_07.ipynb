```python
import pandas as pd

df = pd.DataFrame({
    "Jam_Belajar": [1, 2, 3, 4, 5],
    "Nilai": [55, 65, 75, 85, 95]
})

df
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Jam_Belajar"]]
y = df["Nilai"]

model = LinearRegression()

model.fit(X, y)

model.predict([[6]])
```

```python
print("Intercept:", model.intercept_)
print("Koefisien:", model.coef_[0])
```

## Kesimpulan

Regresi linier digunakan untuk memprediksi nilai berdasarkan hubungan antara variabel input dan output.
